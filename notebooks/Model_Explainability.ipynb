```python
# Import necessary libraries
import subprocess
import matplotlib.pyplot as plt
import IPython.display as display
import os

# Define a function to run the model explainability script
def run_model_explainability():
    try:
        # Adjust the path to the script based on your directory structure
        result = subprocess.run(['python', '../scripts/Model_Explanaibility.py'], capture_output=True, text=True)

        # Print the output from the script
        print(result.stdout)
        print(result.stderr)
        
        # Check if the plots were saved
        plot_files = ['shap_summary_plot.png', 'shap_force_plot.png', 'shap_dependence_plot.png', 'lime_feature_importance_plot.png']
        for plot_file in plot_files:
            if os.path.exists(plot_file):
                print(f"{plot_file} has been saved.")
                display.display(display.Image(filename=plot_file))  # Display the saved plot

    except Exception as e:
        print(f"An error occurred: {e}")

# Run the model explainability script
run_model_explainability()
```
